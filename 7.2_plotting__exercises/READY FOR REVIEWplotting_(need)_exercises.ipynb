### Exercise 1

```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=.25, yerr=y_err,
             color='#8f0909', linestyle='--',
             marker='o', markerfacecolor='#efde04',
             ecolor='#000000',capsize=5,
             label="Y values with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc="lower right")
plt.show()
```

### Exercise 2

```python
x=np.linspace(0,50,1000)
y=[]

for i in range(len(x)):
    num=math.e**(-x[i]/10)*math.sin(x[i])
    y.append(num)

plt.plot(x,y,color='#2596be')
plt.xlabel('x',color='darkorange',fontsize=15)
plt.ylabel('y',color='darkorange',fontsize=15,rotation=0)
plt.title('$y=e^{(-x/10)}sin(x)$',fontsize=20)
plt.ylim(-1.0,1.0)
plt.grid(alpha=.25)
plt.show()
```
